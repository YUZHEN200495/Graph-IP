```python
import random
import time
import networkx as nx
import matplotlib.pyplot as plt

def generate_network_traffic(num_packets):
    graph = nx.Graph()
    
    for _ in range(num_packets):
        source_ip = ".".join(str(random.randint(0, 255)) for _ in range(4))
        destination_ip = ".".join(str(random.randint(0, 255)) for _ in range(4))
        packet_size = random.randint(1, 1024)
        
        # Add nodes and edges to the graph
        graph.add_node(source_ip)
        graph.add_node(destination_ip)
        graph.add_edge(source_ip, destination_ip, packet_size=packet_size)
        
        # Simulate the time interval for traffic generation (in seconds)
        time.sleep(random.uniform(0.1, 1.0))
    
    return graph

def visualize_graph(graph):
    pos = nx.spring_layout(graph)
    edge_labels = nx.get_edge_attributes(graph, 'packet_size')
    
    nx.draw(graph, pos, with_labels=True, font_weight='bold', node_color='skyblue', node_size=800, edge_color='gray')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
    plt.show()

if __name__ == "__main__":
    try:
        num_packets = 10  # You can adjust the number of packets generated
        network_graph = generate_network_traffic(num_packets)
        visualize_graph(network_graph)
    except KeyboardInterrupt:
        print("Stopped generating network traffic data.")
```
